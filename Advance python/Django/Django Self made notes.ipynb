## How to start?

Start Project
* django-admin startproject projectname

Start Application
* python manage.py startapp appname

Add this application to the project in settings.py file
* INSTALLED_APPS = [

    'appname'
    
   ]

define view function inside views.py
* from django.http import HttpResponse

* def hellowview(request):

      return HttpResponse('hellow world')
      

define url patterns for our view function inside urls.py
* from appname import views
* urlpatterns=[

     url(r'^hello/',views.helloview),
     
     ]
     
     or
     
* from django.urls import path,include
* urlpatterns = [

    path('hello/', views.hellowview, name='hellowview'),
    
    path('admin/', admin.site.urls),
   ]

runserver
* python manage.py runserver

send request
* http://127.0.0.1:8000/hello

* Advantages of defining urls at applilcation level:
   * it promotes reusability of applications
   * Project level urls.py is clean and more readable


Certainly! If you prefer to use include for routing, you can structure your urls.py like this:

from django.contrib import admin

from django.urls import path, include

from appname import views

    path('hello/', include('appname.urls')),
    
    path('admin/', admin.site.urls),
    
]
